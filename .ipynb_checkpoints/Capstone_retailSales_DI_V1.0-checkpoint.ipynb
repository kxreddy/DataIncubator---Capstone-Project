# Train the Model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder,  MinMaxScaler, MaxAbsScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import GridSearchCV
```

```python
df_copy = merged_sales_data_gcpi.copy()
```

```python
df_copy
```

```python
y = df_copy['weekly_sales']
X = df_copy.drop('weekly_sales',axis=1)
```

### Feature engineering 

Utilizing all the features, all numerical features are transformed using standar scaler and categorical features are transformed using OnehotEncoder. 

```python
# Define transformers for different columns
numeric_features = [ 'temperature', 'fuel_price', 'cpi', 'unemployment', 'year', 'gscpi']
categorical_features = ['holiday_flag', 'store', 'month',  'week_day']

numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder())
])

# Combine transformers using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
#feature scaling and one hot code categorical features
scaled_X = preprocessor.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.2, random_state=42)
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
#    mse = mean_squared_error(y_test, y_pred)
#    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    return r2

def evaluate_regressors_r2(regressors, regressor_names, X_train, y_train, X_test, y_test):
    r2s = [evaluate_model(regressor, X_train, y_train, X_test, y_test) for regressor in regressors]
    regressor_r2s = dict(zip(regressor_names, r2s))
    df = pd.DataFrame.from_dict(regressor_r2s, orient='index')
    df = df.reset_index()
    df.columns = ['regressor_name', 'R2']
    return df.sort_values('R2', ignore_index=True)
```

```python
alpha_values = [0.1, 1.0, 10.0,15.0]  # Adjust this list based on your domain knowledge
param_grid = {'alpha': alpha_values}

linear_regressor = LinearRegression()
polynomial_features = PolynomialFeatures(degree=2)
polynomial_regressor = Pipeline([("polynomial_features", polynomial_features),
                                 ("linear_regression", linear_regressor) ])
#ridge_regressor = Ridge()
ridge_regressor = GridSearchCV(Ridge(), param_grid, cv=5)
decision_tree_regressor = DecisionTreeRegressor()  
random_forest_regressor = RandomForestRegressor(n_estimators=200, random_state=42)
knn_regressor = KNeighborsRegressor(n_neighbors=5, weights='uniform')
```

```python
#  list of regressors
regressors = [linear_regressor, polynomial_regressor, ridge_regressor, 
              decision_tree_regressor, random_forest_regressor, knn_regressor]
#  names of regressors
regressor_names = ["Linear Regression", "Polynomial Regression", "Ridge Regression", "Decision Tree Regression", 
                   "Random Forest Regression", "K-Nearest Neighbour Regression"]
```

```python
evaluate_regressors_r2(regressors, regressor_names, X_train, y_train, X_test, y_test)
```

```python
polynomial_regressor.fit(X_train, y_train)
```

```python
y_pred = polynomial_regressor.predict(X_test)
```

```python
print(('prediction = ' +
       f'{linear_regressor.intercept_} +\n' +
       ' +\n'.join([f'{n} * {f}' for f, n in zip(numeric_features, linear_regressor.coef_)])))
```

```python
print(('prediction = ' +
       f'{linear_regressor.intercept_} +\n' +
       ' +\n'.join([f'{n} * {f}' for f, n in zip(categorical_features, linear_regressor.coef_)])))
```

```python
# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2):", r2)
```

### Conclusion :
The analysis found that sales during holiday weeks are significantly higher than during non-holiday weeks, with sales almost doubling on an average. Additionally, there is a strong seasonal component to the sales data. The average sales of the top performing stores are up  ~ 3 to 4 times higher than the lowest performing stores. And , there is no significant correlation between sales and GSCPI has been observed.!

The best model for predicting future sales is found to be LinearRegression model with features transformed to polynomial features with 2nd degree, which achieved an R2 of 0.9626.
